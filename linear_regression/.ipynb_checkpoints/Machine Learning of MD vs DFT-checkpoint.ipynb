##Compare the DFT results to the Molecular Dynamics results to find the optimized parameter in the MD force field for aluminum oxide surface

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

###Each trials have different epsilon vaule of Al and O atom for MD force field.  VASP is DFT results.

```python
data=pd.read_csv('MD_VASP.csv', index_col=0)
data
```

Make some plots

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
for i in range(1, 4):
    _ = data.plot(x='trial%s' % (i), y='VASP', kind='scatter', sharey=True,
                  ax=axes[0, i-1], figsize=[10, 10])
for i in range(4, 7):
    _ = data.plot(x='trial%s' % (i), y='VASP', kind='scatter', sharey=True,
                  ax=axes[1, i-4], figsize=[10, 10])
for i in range(7, 10):
    ax = data.plot(x='trial%s' % (i), y='VASP', kind='scatter', sharey=True,
                   ax=axes[2, i-7], figsize=[10, 10])
    ax.set_xlabel("MD")
```

###Find linear regression model for each plot

```python
from scipy.optimize import fmin
```

```python
ss = lambda theta, x, y: np.sum((y - theta[0] - theta[1]*x) ** 2)
```

```python
X = []
for i in np.arange(1, 10):
    datavalue = data['trial%s' % (i)].values
    X.append(datavalue)

Y = data['VASP'].values
```

```python
b0 = []
b1 = []
for i in range(len(X)):
    param = fmin(ss, [0, 1], args=(X[i], Y))
    b0.append(param[0])
    b1.append(param[1])
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
for i in range(0, 3):
    _ = data.plot(x='trial%s' % (i+1), y='VASP', kind='scatter', sharey=True,
                  ax=axes[0, i], figsize=[10, 10])
    ax = axes[0, i]
    ax.plot([0, 60], [b0[i], b0[i]+b1[i]*60], 'r')
    label = 'm={0:.2f}'.format(b1[i])
    ax.annotate(label, xy=(45, 20))
for i in range(3, 6):
    _ = data.plot(x='trial%s' % (i+1), y='VASP', kind='scatter', sharey=True,
                  ax=axes[1, i-3], figsize=[10, 10])
    ax = axes[1, i-3]
    ax.plot([0, 60], [b0[i], b0[i]+b1[i]*60], 'r')
    label = 'm={0:.2f}'.format(b1[i])
    ax.annotate(label, xy=(45, 20))
for i in range(6, 9):
    _ = data.plot(x='trial%s' % (i+1), y='VASP', kind='scatter', sharey=True,
                  ax=axes[2, i-6], figsize=[10, 10])
    ax = axes[2, i-6]
    ax.plot([0, 60], [b0[i], b0[i]+b1[i]*60], 'r')
    label = 'm={0:.2f}'.format(b1[i])
    ax.annotate(label, xy=(45, 20))
    ax.set_xlabel("MD")
```
